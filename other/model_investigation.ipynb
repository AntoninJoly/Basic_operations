```python
import xgboost as xgb
import sys
import os
import matplotlib.pyplot as plt
import pickle

model_dir = '../model'
list_model = [os.path.join(model_dir, i) for i in os.listdir(model_dir)]
print(list_model)
print(xgb.__version__)
```

```python
fig, ax = plt.subplots(figsize=(10, 20))
model = pickle.load(open(list_model[0], 'rb'))
xgb.plot_importance(model, ax=ax)
plt.title(list_model[0])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 7))
model = pickle.load(open(list_model[1], 'rb'))
xgb.plot_importance(model, ax=ax)
plt.title(list_model[1])
plt.show()
```
