```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import sys
import base64
import json
import codecs
import pickle

json_file = 'C:/Users/PAC/Desktop/image2020-02-18 01_27_48.370275.json'
img_file = 'C:/Users/PAC/Desktop/Project/ELProject/data/Bottle_25cl/image2020-02-18 01_27_48.370275.jpg'
```

```python
f = open(json_file,)
data = json.load(f)

im_bytes = base64.b64decode(data['imageData'])
im_arr = np.frombuffer(im_bytes, dtype=np.uint8)  # im_arr is one-dim Numpy array
img = cv2.imdecode(im_arr, flags=cv2.IMREAD_COLOR)
```

```python
plt.imshow(img)
```
