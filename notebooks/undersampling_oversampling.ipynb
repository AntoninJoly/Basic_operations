```python
import numpy as np
import matplotlib.pyplot as plt

from imblearn.under_sampling import RandomUnderSampler, ClusterCentroids
from imblearn.over_sampling import SMOTE, SMOTENC

np.random.seed(0)
```

# Random image generation and associated label for multiclass classification

```python
n, nx, ny, c = 100, 192, 192, 3

X = np.random.uniform(low=0, high=255, size=(n, nx, ny, 3))
y = np.random.randint(3, size=n)

print(X.shape, y.shape)
```

# Undersampling

```python
rus = RandomUnderSampler(random_state=0, replacement=True)
X_us, y_us = rus.fit_resample(X.reshape((n, nx*ny*c)), y)
X_us = X_rus.reshape(-1, nx, ny, c)

# cc = ClusterCentroids(random_state=0)
# X_rus, y_rrus = cc.fit_resample(X.reshape((n, nx*ny*c)), y)
# X_rus = X_rus.reshape(-1, nx, ny, c)

_, count = np.unique(y, return_counts = True)
_, count_us = np.unique(y_us, return_counts = True)
```

# Visualization

```python
plt.figure(figsize=(20,7))
plt.subplot(1,2,1) 
plt.bar(np.unique(y), count)
plt.subplot(1,2,2) 
plt.bar(np.unique(y), count_us)

plt.tight_layout()
plt.show()
```

# Oversampling

```python
n, n_features = 1000, 15

X = np.random.uniform(low=0, high=255, size=(n, n_features))
y = np.random.randint(5, size=n)

print(X.shape, y.shape)
```

```python
sm = SMOTE(random_state=42)
X_os, y_os = sm.fit_resample(X, y)

_, count = np.unique(y, return_counts = True)
_, count_os = np.unique(y_os, return_counts = True)
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1,2,1) 
plt.bar(np.unique(y), count)
plt.subplot(1,2,2) 
plt.bar(np.unique(y), count_os)

plt.tight_layout()
plt.show()
```
